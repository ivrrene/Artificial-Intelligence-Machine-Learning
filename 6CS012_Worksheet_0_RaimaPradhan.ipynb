### Task 1: Unit Converter
> Run this cell. It will prompt you to choose a conversion type and enter a value.

````python
def convert_length(value, direction):
    if direction == '1':
        result = value * 3.28084
        print(f"  Result: {value} m = {result:.4f} ft")
    elif direction == '2':
        result = value / 3.28084
        print(f"  Result: {value} ft = {result:.4f} m")
    else:
        raise ValueError("Invalid direction. Choose '1' or '2'.")
    return result


def convert_weight(value, direction):
    if direction == '1':
        result = value * 2.20462
        print(f"  Result: {value} kg = {result:.4f} lbs")
    elif direction == '2':
        result = value / 2.20462
        print(f"  Result: {value} lbs = {result:.4f} kg")
    else:
        raise ValueError("Invalid direction. Choose '1' or '2'.")
    return result


def convert_volume(value, direction):
    if direction == '1':
        result = value * 0.264172
        print(f"  Result: {value} L = {result:.4f} gal")
    elif direction == '2':
        result = value / 0.264172
        print(f"  Result: {value} gal = {result:.4f} L")
    else:
        raise ValueError("Invalid direction. Choose '1' or '2'.")
    return result


def unit_converter():
    print("=" * 40)
    print("         UNIT CONVERTER")
    print("=" * 40)
    print("Choose conversion type:")
    print("  1. Length  (m <-> ft)")
    print("  2. Weight  (kg <-> lbs)")
    print("  3. Volume  (L <-> gal)")

    try:
        conv_type = input("\nEnter choice (1/2/3): ").strip()

        if conv_type == '1':
            print("\n  Length options:")
            print("    1. Meters  ->  Feet")
            print("    2. Feet    ->  Meters")
            direction = input("  Enter direction (1/2): ").strip()
            value = float(input("  Enter value: "))
            return convert_length(value, direction)

        elif conv_type == '2':
            print("\n  Weight options:")
            print("    1. Kilograms  ->  Pounds")
            print("    2. Pounds     ->  Kilograms")
            direction = input("  Enter direction (1/2): ").strip()
            value = float(input("  Enter value: "))
            return convert_weight(value, direction)

        elif conv_type == '3':
            print("\n  Volume options:")
            print("    1. Liters   ->  Gallons")
            print("    2. Gallons  ->  Liters")
            direction = input("  Enter direction (1/2): ").strip()
            value = float(input("  Enter value: "))
            return convert_volume(value, direction)

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
            return None

    except ValueError as e:
        print(f"Error: Invalid input — {e}")
        return None

unit_converter()
````

### Task 2: Mathematical Operations on a List of Numbers
> Run this cell. It will prompt you to choose an operation and enter numbers.

````python
def compute_sum(numbers):
    if not numbers:
        raise ValueError("List is empty. Please provide at least one number.")
    total = 0
    for n in numbers:
        total += n
    return total


def compute_average(numbers):
    if not numbers:
        raise ValueError("List is empty. Please provide at least one number.")
    return compute_sum(numbers) / len(numbers)


def compute_maximum(numbers):
    if not numbers:
        raise ValueError("List is empty. Please provide at least one number.")
    max_val = numbers[0]
    for n in numbers:
        if n > max_val:
            max_val = n
    return max_val


def compute_minimum(numbers):
    if not numbers:
        raise ValueError("List is empty. Please provide at least one number.")
    min_val = numbers[0]
    for n in numbers:
        if n < min_val:
            min_val = n
    return min_val


def math_operations():
    print("=" * 40)
    print("      MATHEMATICAL OPERATIONS")
    print("=" * 40)
    print("Choose an operation:")
    print("  1. Sum")
    print("  2. Average")
    print("  3. Maximum")
    print("  4. Minimum")

    try:
        choice = input("\nEnter choice (1/2/3/4): ").strip()
        raw = input("Enter numbers separated by spaces (e.g. 3 7 2 9 4): ").strip()

        if not raw:
            raise ValueError("No numbers were entered.")

        numbers = [float(x) for x in raw.split()]
        print(f"\nNumbers entered: {numbers}")

        if choice == '1':
            result = compute_sum(numbers)
            print(f"Sum     = {result}")
        elif choice == '2':
            result = compute_average(numbers)
            print(f"Average = {result:.4f}")
        elif choice == '3':
            result = compute_maximum(numbers)
            print(f"Maximum = {result}")
        elif choice == '4':
            result = compute_minimum(numbers)
            print(f"Minimum = {result}")
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 4.")
            return None

        return result

    except ValueError as e:
        print(f"Error: Invalid input — {e}")
        return None

math_operations()
````

## 4.3 Exercise on Nested Lists
> Each cell defines a function and tests it with the example from the worksheet.

#### 1. Flatten a Nested List

````python
def flatten(lst):
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result


# Example
inp = [[1, 2], [3, 4], [5]]
print(f"Input : {inp}")
print(f"Output: {flatten(inp)}")   # Expected: [1, 2, 3, 4, 5]
````

#### 2. Accessing Nested List Elements

````python
def access_nested_element(lst, indices):
    element = lst
    for idx in indices:
        element = element[idx]
    return element
lst     = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
print(f"Input : lst={lst}, indices={indices}")
print(f"Output: {access_nested_element(lst, indices)}")   # Expected: 6
````

#### 3. Sum of All Elements in a Nested List

````python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

inp = [[1, 2], [3, [4, 5]], 6]
print(f"Input : {inp}")
print(f"Output: {sum_nested(inp)}")   # Expected: 21
````

#### 4. Remove Specific Element from a Nested List

````python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result
lst  = [[1, 2], [3, 2], [4, 5]]
elem = 2
print(f"Input : lst={lst}, elem={elem}")
print(f"Output: {remove_element(lst, elem)}")
````

#### 5. Find the Maximum Element in a Nested List

````python
def find_max(lst):
    max_val = None
    for item in lst:
        candidate = find_max(item) if isinstance(item, list) else item
        if max_val is None or candidate > max_val:
            max_val = candidate
    return max_val
inp = [[1, 2], [3, [4, 5]], 6]
print(f"Input : {inp}")
print(f"Output: {find_max(inp)}")   # Expected: 6
````

#### 6. Count Occurrences of an Element in a Nested List

````python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count
lst  = [[1, 2], [2, 3], [2, 4]]
elem = 2
print(f"Input : lst={lst}, elem={elem}")
print(f"Output: {count_occurrences(lst, elem)}")
````

#### 7. Flatten a List of Lists of Lists (any depth)

````python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

inp = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(f"Input : {inp}")
print(f"Output: {deep_flatten(inp)}")
````

#### 8. Nested List Average

````python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)
inp = [[1, 2], [3, 4], [5, 6]]
print(f"Input : {inp}")
print(f"Output: {average_nested(inp)}")
````

### Problem 4: Matrix Operations

````python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("A =")
print(A)
print("B =")
print(B)
print()

A_inv = np.linalg.inv(A)
A_times_Ainv = np.dot(A, A_inv)

print("PROOF 1: A · A⁻¹ = I")
print("A⁻¹ =")
print(A_inv)
print("A · A⁻¹ =")
print(np.round(A_times_Ainv, decimals=10))
print("Confirmed (np.allclose):", np.allclose(A_times_Ainv, np.eye(2)))
print()

AB = np.dot(A, B)
BA = np.dot(B, A)

print("PROOF 2: AB ≠ BA")
print("AB =")
print(AB)
print("BA =")
print(BA)
print("AB == BA?", np.array_equal(AB, BA), "  →  Matrix multiplication is NOT commutative.")
print()

AB_T  = AB.T
BT_AT = np.dot(B.T, A.T)

print("PROOF 3: (AB)ᵀ = BᵀAᵀ")
print("(AB)ᵀ =")
print(AB_T)
print("Bᵀ · Aᵀ =")
print(BT_AT)
print("Equal:", np.array_equal(AB_T, BT_AT))
````

#### Solving the System of Linear Equations
```
2x - 3y + z  = -1
 x -  y + 2z = -3
3x +  y - z  =  9
```

````python
A_sys = np.array([[ 2, -3,  1],
                  [ 1, -1,  2],
                  [ 3,  1, -1]], dtype=float)

B_sys = np.array([-1, -3, 9], dtype=float)

print("System of equations:  AX = B")
print("A =")
print(A_sys)
print("B =", B_sys)
print()

A_sys_inv = np.linalg.inv(A_sys)
X_inv = np.dot(A_sys_inv, B_sys)

print("METHOD 1 — Inverse method (X = A⁻¹ · B):")
print(f"  x = {X_inv[0]:.4f}")
print(f"  y = {X_inv[1]:.4f}")
print(f"  z = {X_inv[2]:.4f}")
print()

X_solve = np.linalg.solve(A_sys, B_sys)

print("METHOD 2 — np.linalg.solve:")
print(f"  x = {X_solve[0]:.4f}")
print(f"  y = {X_solve[1]:.4f}")
print(f"  z = {X_solve[2]:.4f}")
print()

print("Verification — A · X_solve:")
print(" ", np.round(np.dot(A_sys, X_solve), 6))
print("Expected B:")
print(" ", B_sys)
````

## 10.2 Experiment: How Fast is NumPy?
> Run each cell to measure and compare Python vs NumPy performance.

````python
import time
import numpy as np

SIZE = 1_000_000

py_a = list(range(SIZE))
py_b = list(range(SIZE))
np_a = np.arange(SIZE, dtype=float)
np_b = np.arange(SIZE, dtype=float)

print(f"Data prepared: two arrays/lists of size {SIZE:,} each.")
````

````python
print("=" * 50)
print("1. ELEMENT-WISE ADDITION  (size = 1,000,000)")
print("=" * 50)
start = time.time()
result_py = [py_a[i] + py_b[i] for i in range(SIZE)]
py_time = time.time() - start
print(f"  Python List : {py_time:.5f} seconds")

start = time.time()
result_np = np_a + np_b
np_time = time.time() - start
print(f"  NumPy Array : {np_time:.5f} seconds")
print(f"  Speedup     : {py_time / np_time:.1f}x faster with NumPy")
````

````python
print("=" * 50)
print("2. ELEMENT-WISE MULTIPLICATION  (size = 1,000,000)")
print("=" * 50)

start = time.time()
result_py = [py_a[i] * py_b[i] for i in range(SIZE)]
py_time = time.time() - start
print(f"  Python List : {py_time:.5f} seconds")

start = time.time()
result_np = np_a * np_b
np_time = time.time() - start
print(f"  NumPy Array : {np_time:.5f} seconds")
print(f"  Speedup     : {py_time / np_time:.1f}x faster with NumPy")
````
